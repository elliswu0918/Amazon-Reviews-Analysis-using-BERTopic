```python
import pandas as pd
df=pd.read_csv("amazon_music_reviews_with_sentiment.csv", encoding='ISO-8859-1')
df
```

```python
df['word_count'] = df['cleaned_text'].astype(str).apply(lambda x: len(x.split()))
a1=df['word_count']>=10
new_data=df[a1]
new_data
```

```python
import statsmodels.api as sm
import pandas as pd


# 2. 選擇 X 與 Y
X = new_data[['price', 'rating', 'sentiment_score', 'Topic 1', 'Topic 2', 'Topic 3', 'Topic 4', 'word_count']]
Y = new_data['rank']

# 3. 確保所有變數為數值型
X = X.apply(pd.to_numeric, errors='coerce')
Y = pd.to_numeric(Y, errors='coerce')

# 4. 移除缺失值（NaN）資料列
df_clean = pd.concat([X, Y], axis=1).dropna()
X = df_clean.drop(columns=['rank'])
Y = df_clean['rank']

# 5. 加入常數項（截距）
X = sm.add_constant(X)

# 6. 建立並擬合模型
model = sm.OLS(Y, X).fit()

# 7. 印出完整回歸表
print(model.summary())
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
new_data_clean['log_word_count'] = np.log1p(new_data_clean['word_count'])  # 解決偏態
new_data_clean['scaled_price'] = StandardScaler().fit_transform(new_data_clean[['price']])
```

```python
new_data_clean['rank_score'] = 1 - (new_data_clean['rank'] - 1) / 99
```

```python
#有主題
X = new_data_clean[['scaled_price', 'rating', 'sentiment_score','Topic 1', 'Topic 2', 'Topic 3', 'Topic 4', 'log_word_count']]
Y = new_data_clean['rank_score']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#無主題
non_X = new_data_clean[['scaled_price', 'rating', 'sentiment_score', 'log_word_count']]
non_Y = new_data_clean['rank_score']
non_X_train,non_X_test,non_Y_train,non_Y_test=train_test_split(non_X,non_Y,test_size=0.2,random_state=42)
```

```python
import statsmodels.api as sm

# 在 X_train 上添加截距項
X_train_with_intercept = sm.add_constant(X_train)

# 使用 statsmodels 進行線性回歸
model = sm.OLS(Y_train, X_train_with_intercept).fit()

# 預測
y_pred_stats = model.predict(sm.add_constant(X_test))

# 評估模型
mse_stats = mean_squared_error(Y_test, y_pred_stats)
r2_stats = r2_score(Y_test, y_pred_stats)

print(f"Statsmodels Linear Regression MSE: {mse_stats}")
print(f"Statsmodels Linear Regression R^2: {r2_stats}")

# 打印詳細的回歸結果
print(model.summary())
```
