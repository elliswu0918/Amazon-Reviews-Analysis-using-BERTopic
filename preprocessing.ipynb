```python
import pandas as pd
df = pd.read_excel('meta_Musical_Instruments_1.xlsx')
df
```

```python
df1 = pd.read_excel('Rank 前 100個產品品項.xlsx')
df1 
```

```python
df2 = pd.read_excel('review_Musical_Instrument_1.xlsx')
df3 = pd.read_excel('review_Musical_Instrument_2.xlsx')
df4 = pd.read_excel('review_Musical_Instrument_3.xlsx')
ct=pd.concat([df2,df3,df4])
ct
```

```python
df_1=df1.merge(ct,how='left')
df_1=df_1.drop_duplicates('asin')
df_1=df_1[['asin','parent_asin','rank']]
df_1
```

```python
dfdf=df_1.merge(df,on="parent_asin",how="left")
dfdf=dfdf[['asin','average_rating','price']]
dfdf['average_rating'] = dfdf['average_rating'].fillna(dfdf['average_rating'].mean())
dfdf['price'] = dfdf['price'].fillna(dfdf['price'].mean())
dfdf
```

```python
all_data=dfdf.merge(ct,on="asin",how='left')
all_data=all_data.merge(df_1,on='asin')
all_data
```

```python
dfdf.to_excel('100產品詳細資訊.xlsx', sheet_name='工作表1', index=False)
```

```python
all_data.to_excel('all_data.xlsx', sheet_name='工作表1', index=False)
```
